```python
from PIL import Image
import PIL
import numpy as np
import matplotlib.pyplot as plt

#encrypt:
yaya = Image.open('yaya.jpg')
yaya = np.array(yaya)

insert = Image.open('letter.jpg')
insert = insert.resize((768, 768))
insert = np.array(insert)

insert[insert<128] = 1
insert[insert>128] = 0

even = yaya%2==0
yaya[even] += 1
new_yaya = yaya + insert

#decode
hidden = new_yaya%2*255

plt.subplot(121)
plt.imshow(new_yaya)
plt.title('Encrypted Image')
plt.subplot(122)
plt.imshow(hidden)
plt.title('Hidden Image')
```
